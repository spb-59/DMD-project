```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
SR = pd.read_csv('features2/SR.csv')
AFIB = pd.read_csv('features2/AFIB.csv')

# Check if both DataFrames have the same number of features
assert SR.shape[1] == AFIB.shape[1], "DataFrames must have the same number of features."

# Number of features
num_features = SR.shape[1]

# Create subplots
fig, axes = plt.subplots(nrows=9, ncols=6, figsize=(20, 30))  # Adjust rows and cols based on number of features
axes = axes.flatten()  # Flatten the 2D array of axes

for i in range(num_features):
    axes[i].scatter(SR.iloc[:, i], AFIB.iloc[:, i], alpha=0.5)
    axes[i].set_title(f'Feature {i+1}')
    axes[i].set_xlabel('SR')
    axes[i].set_ylabel('AFIB')

# Hide any unused subplots
for j in range(num_features, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
